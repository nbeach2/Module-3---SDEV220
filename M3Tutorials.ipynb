```python
class Solution:
    def sort012(self, arr):
        Amount_0 = 0
        Amount_1 = 0
        Amount_2 = 0
        
        for x in arr:
            if x == 0:
                Amount_0 += 1
            elif x == 1:
                Amount_1 += 1
            else:
                Amount_2 += 1
                
        i = 0
        
        for _ in range(Amount_0):
            arr[i] = 0
            i += 1
        for _ in range(Amount_1):
            arr[i] = 1
            i += 1
        for _ in range(Amount_2):
            arr[i] = 2
            i += 1
arr = [0, 1, 1, 0, 1, 2, 0, 1, 2]
Solution().sort012(arr)
print(arr)
```

```python
class Solution:
    def binarysearch(self, arr, k):
        index = 0
        
        for x in arr:
            if x == k:
                return index
            index += 1
        return -1

arr = [1, 2, 3, 4, 5]
k = 4
obj = Solution()
output = obj.binarysearch(arr, k)
print(output)
arr = [1, 2, 3, 4, 5]
k = 145
output = obj.binarysearch(arr, k)
print(output)
```
